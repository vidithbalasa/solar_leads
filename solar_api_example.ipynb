```python
import requests
import os
```

```python
SOLAR_API_KEY = os.environ.get('GOOGLE_MAPS_API_KEY')
URL = f"https://solar.googleapis.com/v1/buildingInsights:findClosest?key={SOLAR_API_KEY}"
```

```python
# Get latitude and longitude from user
```

```python
addy = "2415 E Pinehurst Ave, Fresno, CA 93730"
response = requests.get(f"https://maps.googleapis.com/maps/api/geocode/json?address={addy}&key={SOLAR_API_KEY}")
data = response.json()
location = data['results'][0]['geometry']['location']
params = {
    'location.latitude': location['lat'],
    'location.longitude': location['lng'],
    'requiredQuality': 'HIGH'
}
```

```python
response = requests.get(URL, params=params)
response.json()
```
